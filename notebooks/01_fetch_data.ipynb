```python
import requests
import pandas as pd
import numpy as np
import time
import os
from dotenv import load_dotenv
load_dotenv()
TOKEN = os.getenv("GITHUB_TOKEN")
HEADERS = {"Authorization": f"token {TOKEN}"}

# Test token
r = requests.get("https://api.github.com/user", headers=HEADERS)
print(r.json()["login"])
```

```python
import ijson
import pandas as pd

records = []
MAX_REPOS = 2000

with open("../data/repo_metadata.json", "r", encoding="utf-8") as f:
    for item in ijson.items(f, "item"):
        
        if item.get("isArchived"):
            continue
        if item.get("isFork"):
            continue
        if item.get("forks", 0) < 5:
            continue

        records.append({
            "full_name":   item.get("nameWithOwner"),
            "owner":       item.get("owner"),
            "name":        item.get("name"),
            "stars":       item.get("stars", 0),
            "forks":       item.get("forks", 0),
            "issues":      item.get("issues", 0),
            "description": item.get("description"),
            "license":     item.get("license"),
            "pushedAt":    item.get("pushedAt"),
            "defaultBranchCommitCount": item.get("defaultBranchCommitCount", 0),
        })

        if len(records) >= MAX_REPOS:
            break

df_kaggle = pd.DataFrame(records)
print(f"Total repo: {len(df_kaggle)}")
print(df_kaggle.head())
```

```python
df_kaggle.to_csv("../data/kaggle_filtered.csv", index=False)
print("Tersimpan ke kaggle_filtered.csv!")
```
